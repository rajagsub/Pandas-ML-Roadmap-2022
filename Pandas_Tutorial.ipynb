### Now lets try to Create a function and get the details

```python
def run_scored_against(batsman_name):
    mask2 = delivery['batsman'] == batsman_name
    vk1 = delivery[mask2]
    grp = vk1.groupby('bowling_team')
    return grp['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

## The isin() function
- find the most destructive death over batsman in the history of IPL 
- strike rate = (number of runs/number of balls)*100
- min batsman 200 balls in over 16-20.


```python
#Filter deathover from DF
death_over = delivery[delivery['over']>15]
```

```python
# Here we count the number of balls played by batsman in last 16-20 overs
balls_all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
# here we are assigning the condition to the series which will retun the boolean result, once we pass it the series again 
# with an index it will get us the values for the index.
cond = balls_all_batsman > 200
batsman_list = balls_all_batsman[cond].index.tolist()
```

```python
# Runs scored by all the above batsman 
# Balls played by all the above batsman

# We are checking the above batsman list who hit more than 200 balls in last 16-20 overs in the actual dataset and filter them
final_df  = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
# Now lets groupby by batsman and get the sum of runs scored by them
runs = final_df.groupby('batsman')['batsman_runs'].sum().sort_values()
```

```python
# Now lets groupby by batsman and get the Count of balls he played
balls = final_df.groupby('batsman')['batsman_runs'].count().sort_values()
```

```python
strike_rate = (runs/balls) * 100
```

```python
strike_rate.sort_values(ascending=False).head(1).index[0]
```

## Merge Function = = Joins
- Using an usecase, find out the orange cap holder for each season. Using matches.csv and delivery.csv
- id is same in both the files.

```python
data.head(1)
```

```python
delivery.head(1)
```

```python
# Merge 
merge_new = delivery.merge(data,left_on='match_id',right_on='id')
merge_new.head(1)
```

```python
print(data.shape)
print(delivery.shape)
print(merge_new.shape)
```

```python
# 1. Groupby on season and batsman
# 2. get the sum values from the groupby
# 3. Sort the values in descending order (to get highest runs on each season)
# 4. Convert this to the DF using reset index()
# 5. Drop duplicates subset on seasons and keep the first entry for each season 
# 6. Sort the Dataframe based on seasons (ascending)
# 7. Just hide the runs if req 

merge_new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
```

## Pandas Pivot Table
- To determine IPL team on which over they hit more sixes 
- Components of pivot table: 
   index --> grouping clause 
   columns --> wht columns you needed after grouping
   values --> on which numric field u need to perform pivot 
   aggfunc --> what stastical function you need to perform  

```python
data.head(2)
```

```python
mask = delivery['batsman_runs'] == 6
six = delivery[mask]
six.head(2)
```

```python
#Preparing Pivot

pt = six.pivot_table(index = 'over', columns= 'batting_team', values = 'batsman_runs', aggfunc= 'count')
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
